```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print("Modules Imported")
```

```python
#Reading dataset
corona_dataset=pd.read_csv("covid19_Confirmed_dataset.csv")
corona_dataset.head(10)
```

```python
#to find the size of dataset we use shape
corona_dataset.shape
```

```python
#removing unwanted column
# here inplace used so that we need not assign this dataset to another variable (for eg:df)
#df=corona_dataset.drop(["Lat","Long"],axis=1)--.here axis indicated the column(1) if we want row to be deleted we can use(axis=0).
#df.head() here when we accidently  call corona_dataset.head()-->it returns dataframe where lat nd long will still be there. 
corona_dataset.drop(["Lat","Long"],axis=1,inplace=True)
corona_dataset.head(10)
```

```python
#aggregate the rows by country/region (this is like index)
corona_dataset_aggregated=corona_dataset.groupby("Country/Region").sum()
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
#visualizing data related to country for eg:China
corona_dataset_aggregated.loc["China"].plot()
```

```python
#calculating good measure
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["Italy"].plot()
plt.legend()
```

```python
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["Italy"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
plt.legend()
```

```python
#finding the measure of china in first 3 days
corona_dataset_aggregated.loc["China"][:3].plot()
#the graph below shows that in first day there were 550 cases,second day(650),third day (900)
```

```python
#calculating first derivative of the curve
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
#finding maximum infection rate in china,italy and spain
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["Italy"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

```python
#finding maximum infection rate for all of the countries
countries=list(corona_dataset_aggregated.index)
max_infection_rates=[]
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["max_infection_rate"]=max_infection_rates   
```

```python
corona_dataset_aggregated.head()
```

```python
#create a new dataframe with only needed column(ie.we need only max infection rate column)
corona_data=pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
corona_data
```

```python
corona_data.head()
```

```python
#WORLD HAPPINESS REPORT
happiness_report=pd.read_csv("worldwide_happiness_report.csv")
happiness_report.head()
```

```python
#drop the useless column
useless_cols=["Overall rank","Score","Generosity","Perceptions of corruption"]
happiness_report.drop(useless_cols,axis=1,inplace=True)
happiness_report.head()
```

```python
#change indices of the dataframe
happiness_report.set_index("Country or region",inplace=True)
happiness_report.head()
```

```python
#joining both corona data and happiness report data
corona_data.head()
```

```python
happiness_report.head()
```

```python
#perform inner join on 2 datasets
data=corona_data.join(happiness_report,how="inner")
data.head()
```

```python
#CORRELATION MATRIX
data.corr()
```

```python
data.head()
```

```python
#visualizing results
#plotting gdp vs maximum infection rate
x=data["GDP per capita"]
y=data["max_infection_rate"]
sns.scatterplot(x,np.log(y))
```
